# Word Embedding Techniques using Embedding Layer in Keras

```python
import numpy as np
import keras
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['rajendra is a good','deep learning is intersting','we learned well','apple juse is healthy']
```

```python
sent
```

```python
voc_size=10000
```

# one hot representation

```python
onehot_rep=[one_hot(word,voc_size) for word in sent]
```

```python
onehot_rep
```

## Word Embedding Represntation

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
```

```python
sent_length=8
emb_doc=pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
```

```python
emb_doc
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(emb_doc))
```

```python
print(model.predict(emb_doc)[0])
```
